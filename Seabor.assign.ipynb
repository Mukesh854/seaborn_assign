Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:

Use: Scatter plots display individual data points as markers on a two-dimensional plane. They are great for visualizing the relationship between two continuous variables, identifying patterns, and detecting outliers.

Bar Plot:

Use: Bar plots are used to compare categorical data using rectangular bars. They're effective for showing comparisons between different categories or groups. Variations of bar plots include grouped bar plots and stacked bar plots.

Histogram:

Use: Histograms show the distribution of a single continuous variable. They group data into bins and display the frequency of observations in each bin. Histograms are useful for understanding data distribution, detecting skewness, and identifying modes.

Box Plot (Box-and-Whisker Plot):

Use: Box plots visualize the distribution of data and highlight key summary statistics such as median, quartiles, and potential outliers. They are particularly useful for comparing distributions across different categories or groups.

Heatmap:

Use: Heatmaps display a matrix of data as a grid of colored cells, where the color intensity represents the magnitude of the values. Heatmaps are commonly used to visualize correlations between variables in a dataset or to display various forms of data matrices.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri_data=sns.load_dataset("fmri")
fmri_data
```

```python
sns.set(style="whitegrid")
sns.lineplot(data=fmri_data,x="timepoint" , y="signal",hue="event",style="region")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
titanic_data=sns.load_dataset("titanic")
titanic_data
```

```python
sns.boxplot(data=titanic_data , x="pclass",y="age")
```

```python
sns.boxplot(data=titanic_data , x="pclass", y="fare")
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data=sns.load_dataset("diamonds")
diamonds_data
```

```python
sns.histplot(data=diamonds_data , x="price", bins=15, kde=True, hue="cut" )
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data=sns.load_dataset("flights")
pivot_table=flights_data.pivot_table(index="month",columns="year",values="passengers")

sns.heatmap(data=pivot_table, cmap="coolwarm",annot=True,fmt="d")
```
